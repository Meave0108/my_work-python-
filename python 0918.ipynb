# while loop(條件成立:執行，相反則不執行)

```python
s=0
while s<10:
    print(s)
    s=s+1
```

```python
mag="hsdjkahdfj"

while(mag!="bye"):
    mag=input(">> ")
    print("拍拍")
if ("bye" in mag):
    print("bye")
```
