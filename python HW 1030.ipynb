## 作業一:國內歷年乳癌五年存活率(2000-2005)

```python
c=pd.read_csv('cancer.csv')
c
```

```python
c=c.rename(columns={'依年度別分':'year'})
c
```

```python
c=c[c.year>=2000]
```

```python
cs=c.drop(['year','子宮頸癌觀察存活率','子宮頸癌相對存活率','大腸直腸癌觀察存活率','大腸直腸癌相對存活率'],axis=1)
cs.columns = ['observed survival ', 'relative survival ']

year=list(range(2000,2006))
cs.index=year
cs
```

```python

cs.plot(use_index=True,title='The Survival Rate of Breast Cancer',grid=True,colormap='coolwarm',style=['o-','^-'])
```
