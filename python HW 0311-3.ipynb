## 各科級分人數分布折線圖

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a=pd.read_excel('各科級分人數百分比累計表.xls')
a
```

```python
aa=a.iloc[:,[0,1,5,9,13,17]]
aa
aa=aa.drop(index=0)
aa
```

```python
plt.plot(aa['級分'], aa['國文'], linestyle = '-',linewidth = 2, color = '#000000')
plt.plot(aa['級分'], aa['英文'], linestyle = '-',linewidth = 2, color = '#B1091D')
plt.plot(aa['級分'], aa['數學'], linestyle = '-',linewidth =2, color = '#003049')
plt.plot(aa['級分'], aa['社會'], linestyle = '-',linewidth = 2, color = '#497A12')
plt.plot(aa['級分'], aa['自然'], linestyle = '-',linewidth = 2, color = '#96ACB7')
plt.title('GSAT_2019')
plt.xlabel('Grade')
plt.ylabel('number')
plt.legend(('chinese', 'english', 'math','social science','science'), loc='upper right')  
```
