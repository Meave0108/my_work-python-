## 作業:countour 練習

```python
xL=np.array([6032422,107204002])
yL=np.array([2000,108])
X,Y=np.meshgrid(xL,yL)
X
Y
```

```python
Z=np.random.randint(0,7,(2,2))
Z
```

```python
plt.contour(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel(),cmap="autumn")
```

```python
plt.contourf(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel(),cmap="autumn")
```
