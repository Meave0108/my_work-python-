## 資料整理+泡泡圖練習__新北市各區老年人口比例跟所得中位數之關係

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pop2018= pd.read_excel("2018年12月鄉鎮市區單一年齡人口數.xlsx")
pop2018.head()
pop2018=pop2018.drop("COUNTY_ID",axis=1)
pop2018.head()
```

### 人口總數

```python
pop2018['人口總數(男加女)']=pop2018.男+pop2018.女
pop2018.head()
```

```python
pop2018['年齡(new)']=pop2018.年齡+0.5
pop2018=pop2018.drop("年齡",axis=1)

pop2018.head()
```

```python
pop2018_new=pop2018.groupby(['TOWN_ID','COUNTY','TOWN']).sum()
pop2018_new.head()
```

### 老年人口總數

```python
oldpop=pop2018[pop2018['年齡(new)']>65]
oldpop.head()
```

```python
oldpop=oldpop.groupby(['TOWN_ID','COUNTY','TOWN']).sum()
oldpop.head()
oldpop=oldpop.drop(['男','女','年齡(new)'],axis=1)

oldpop=oldpop.rename(columns={'人口總數(男加女)':'老年人口總數(男加女)'})
oldpop.head()
```

### 合併

```python
pop2018_new_1 = pd.merge(pop2018_new, oldpop,left_on=['TOWN_ID','COUNTY','TOWN'],right_on=['TOWN_ID','COUNTY','TOWN'])

pop2018_new_1.head()
```

### 所得中位數(找到的最新資訊為105年)

```python
tax= pd.read_excel("105年綜合所得稅所得總額申報統計_鄉鎮市區.xlsx")
tax.head()
tax=tax[['鄉鎮市區代碼','縣市名稱','鄉鎮市區名稱','中位數']]
tax.columns = ['TOWN_ID','COUNTY','TOWN','中位數']
tax.head()
```

```python
pop2018_new_2 = pd.merge(pop2018_new_1, tax,left_on=['TOWN_ID','COUNTY','TOWN'],right_on=['TOWN_ID','COUNTY','TOWN'])
pop2018_new_2.head()
```

### 老年人口比例
* 老年人口比例=老年人口/人口總數


```python
pop2018_new_2['老年人口比例(%)']=pop2018_new_2['老年人口總數(男加女)']/pop2018_new_2['人口總數(男加女)']*100
pop2018_new_2.head()
```

```python
new_taipei=pop2018_new_2[pop2018_new_2['COUNTY']=='新北市']
new_taipei
```

```python
from bubbly.bubbly import bubbleplot 
from plotly.offline import plot

my_bubbleplot= bubbleplot(dataset=new_taipei, x_column='中位數', y_column='老年人口比例(%)', 
    bubble_column='TOWN',size_column='老年人口總數(男加女)', color_column='TOWN', 
    x_title="所得中位數", y_title="老年人口比例(%)", title='2018年新北市各區老年人口比例跟所得中位數之關係',scale_bubble=3, height=700)
 
plot(my_bubbleplot,config={'scrollzoom': True})
```
