```python
town = gpd.read_file('TOWN_MOI_1080726.shp',encoding='utf-8')
town.head()
```

```python
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt

pylab.rcParams['figure.figsize'] = (170, 170)
town.plot()

```

```python
df= pd.read_csv("enterovirus.csv")
df.head()
df=df.drop(['週','健保就診總人次','年齡別','就診類別'],axis=1)
df.head()
```

```python
df1=df[(df.年>=2015)&(df.年<2019)]
df1=df1.rename(columns={'縣市':'COUNTY'})


df1
```

```python
df1=df1.groupby('COUNTY').sum()
df1=df1.drop(['年'],axis=1)

df1

```

```python

df_new = pd.merge(town,df1,left_on=('COUNTYNAME'),right_on=('COUNTY'))

df_new.head()

```

```python

df_new.plot(column='腸病毒健保就診人次',  k=10,cmap='Reds')
##尚未解決圖中呈現空白的原因
```
