## 梗圖聊天機器人

```python
from IPython.display import Image
p1=Image(filename='./x.png')
p2=Image(filename='./z.jpg')
p3=Image(filename='./y.jpg')
p4=Image(filename='./w.jpg')

a="aa"
print("≧ω≦:Hi!想跟我聊什麼呢?")

while(a!=" "):
    a=input(">> ")
    if ('難過'in a or'傷心' in a or'不順' in a or 'QQ'in a):
        display(p1)
    elif('錢' in a or '借'in a):
        display(p2)
    elif('吃' in a or '餓'in a or '甜點'in a):
        display(p3)
    elif('朋友' in a):
        display(p4)
    elif(a=='bye'):
        break
    else:
        print("?")
```
